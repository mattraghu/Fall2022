```python
from sympy import * 
import pandas as pd
```

```python
#Create a truth table using pandas for: ((NOT A AND B) OR NOT C) XOR (B)

#Define the Dataframe 
df = pd.DataFrame(columns=['A','B','C','NOT A', 'NOT A AND B', 'NOT C', 'NOT C OR (NOT A AND B)', '( NOT C OR (NOT A AND B) ) XOR B'])

#Fill in the Dataframe
for i in range(2):
    for j in range(2):
        for k in range(2):
            df = df.append({'A': i, 'B': j, 'C': k, 'NOT A': not i, 'NOT A AND B': not i and j, 'NOT C': not k, 'NOT C OR (NOT A AND B)': not k or (not i and j), '( NOT C OR (NOT A AND B) ) XOR B': ( (not k or (not i and j)) ^ j )}, ignore_index=True)

#Convert all columns to bools
df = df.astype(bool)

#Print the Dataframe
df
```
