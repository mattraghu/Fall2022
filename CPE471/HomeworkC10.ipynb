```python
#Question 1
#(a) Calculate the maximum space charge width Xar and the maximum space charge
#density Q'»(max) in a MOS capacitor with a p-type silicon substrate at T = 300 K
#for doping concentrations of (i) Na = 7 × 1015 cm-3 and (it) Na = 3 X 1016 cm-3
#(b) Repeat part (a) for T = 350 K.

#(a)
#Constants
e_s = 11.7*8.85e-14 #F/cm
k = 8.617e-5 #eV/K
T = 300 #K
q_e = 1.602e-19 #C

V_t = k*T #eV
print("V_t = ", V_t, "eV")

#(i)
print("--------------------")
print("Values for Na = 7 X 1015 cm-3 and T = 300 K")
Na = 7e15 #cm-3
Ni = 1.5e10 #cm-3

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

x_dt = sqrt(4*o_fp*e_s/( q_e*Na ))
print("x_dt = ", x_dt, "cm")

Q_max = q_e*Na*x_dt
print("Q_max = ", Q_max, "C/cm^2")


#(ii)
print("--------------------")
print("Values for Na = 3e16 cm-3 and T = 300 K")
Na = 3e16 #cm-3

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

x_dt = sqrt(4*o_fp*e_s/( q_e*Na ))
print("x_dt = ", x_dt, "cm")

Q_max = q_e*Na*x_dt
print("Q_max = ", Q_max, "C/cm^2")

#(b)
print("--------------------")
#Constants
T = 350 #K
V_t = k*T #eV
print("V_t = ", V_t, "eV")

#(i)
print("--------------------")
print("Values for Na = 7 X 1015 cm-3 and T = 350 K")
Na = 7e15 #cm-3
Ni = 1.5e10 #cm-3

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

x_dt = sqrt(4*o_fp*e_s/( q_e*Na ))
print("x_dt = ", x_dt, "cm")

Q_max = q_e*Na*x_dt
print("Q_max = ", Q_max, "C/cm^2")


#(ii)
print("--------------------")
print("Values for Na = 3e16 cm-3 and T = 350 K")
Na = 3e16 #cm-3

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

x_dt = sqrt(4*o_fp*e_s/( q_e*Na ))
print("x_dt = ", x_dt, "cm")

Q_max = q_e*Na*x_dt
print("Q_max = ", Q_max, "C/cm^2")
```

```python
#Question 2
#Determine the metal-semiconductor work function difference bs in a MOS
#structure with p-type silicon for the case when the gate is (a) aluminum
#(b) n' polysilicon, and (c) p* polysilicon. Let N, = 6 X 1015 cm-3

#Constants
E_g = 1.12 #eV
q_e = 1.602e-19 #C
V_t = 0.0259 #eV
Ni = 1.5e10 #cm-3

Na = 6e15 #cm-3

#(a)
print("--------------------")
print("Values for Aluminum")
o_m = 3.2
x = 3.25

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

o_ms = o_m - x - o_fp - (E_g/(2))
print("o_ms = ", o_ms, "eV")

#(b)
print("--------------------")
print("Values for n' polysilicon")

o_ms = -(E_g/(2) + o_fp)
print("o_ms = ", o_ms, "eV")

#(c)
print("--------------------")
print("Values for p* polysilicon")

o_ms = (E_g/(2) - o_fp)
print("o_ms = ", o_ms, "eV")

```

```python
#Question 3
#Consider an aluminum gate-silicon dioxide-p-type silicon MOS structure with
#tox = 450 À. The silicon doping is Na = 2 X 101 cm-3 and the flat-band voltage is
#VER = -1.0 V. Determine the fixed oxide charge Q'

#Constants
e_s = 11.7*8.85e-14 #F/cm
V_t = 0.0259 #eV

Na = 2e16 #cm-3
Ni = 1.5e10 #cm-3
V_fb = -1.0 #V
#Convert V_fb to eV
V_fb = V_fb*q_e
tox = 450e-8 #cm

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

o_ms = -.894 #eV

e_ = 3.9*8.85e-14 #F/cm
C_ox = e_/tox
print("C_ox = ", C_ox, "F/cm^2")

Q_ss = -(V_fb-o_ms)*C_ox
print("Q_ss = ", Q_ss, "C/cm^2")
```

```python
#Question 4
#An n polysilicon gate-silicon dioxide-silicon MOS capacitor has an oxide thickness of tox = 18 um = 180 À and a doping of Na = 1015 cm-3. The oxide charge density is O' = 6 X 1010 cm -2. Calculate the (a) flat-band voltage and (b) threshold voltage.

#Constants
e_s = 11.7*8.85e-14 #F/cm
V_t = 0.0259 #eV
q_e = 1.602e-19 #C

Na = 1e15 #cm-3
Ni = 1.5e10 #cm-3
Q_ss = 6e10 #cm-2
tox = 180e-8 #cm

o_fp = V_t*log(Na/Ni)
print("o_fp = ", o_fp, "eV")

E_g = 1.12 #eV
o_ms = -(E_g/(2) + o_fp)
print("o_ms = ", o_ms, "eV")

e_ = 3.97*8.85e-14 #F/cm
C_ox = e_/tox
print("C_ox = ", C_ox, "F/cm^2")

Q_ss = Q_ss * q_e
print("Q_ss = ", Q_ss, "C/cm^2")

o_ms = o_ms * q_e
print("o_ms = ", o_ms, "V")

V_fb = (Q_ss/C_ox) - o_ms
print("V_fb = ", V_fb, "V")

print("--------------------")

x_dt = sqrt(4*o_fp*e_s/( q_e*Na ))
print("x_dt = ", x_dt, "cm")

Q_max = q_e*Na*x_dt
print("Q_max = ", Q_max, "C/cm^2")

V_th= V_fb + (Q_max/C_ox) + 2*o_fp
print("V_th = ", V_th, "V")
```
