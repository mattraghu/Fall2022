```python
from sympy import *
```

```python
#Values

#Effective DoS at 300K
EffectiveDoS = {
    "Silicon" : {
        "N_c" : 2.8e19,#cm^-3
        "N_v" : 1.04e19,
        "E_g" : 1.12
    },
    "Germanium" : {
        "N_c" : 1.04e19,
        "N_v" : 6e18,
        "E_g" : 0.67
    },
    "Silicon Germanium" : {
        "N_c" : 1.5e19, 
        "N_v" : 7.5e18,
        "E_g" : 0.9
    },
    "Gallium Arsenide" : {
        "N_c" : 4.7e17,
        "N_v" : 7e18,
        "E_g" : 1.422
    },


}
```

```python
#Calculate the insintric carrier concentration n at t = 200, 400 and 600k for silicon, germanium and gallium arsenide

#Symbols
n_i = Symbol('n_i')
N_c = Symbol('N_c')
N_v = Symbol('N_v')
E_g = Symbol('E_g')
k = Symbol('k')
T = Symbol('T')

#Eqaution 
IntrinsicCarrierConcentrationInASemiconductorEq = Eq(n_i**2, N_c * N_v * exp(-E_g/(k*T)))

#Constants
k =.0259 #eV/K




#Silicon

#T = 200
T = 200 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens 
Equation = Equation.subs(N_c, EffectiveDoS["Silicon"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Silicon"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Silicon"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
n_iSol = solve(Equation, n_i)[0]

#Print
print("Silicon at T = 200K: " + str(n_iSol) + " cm^-3")



#T = 400
T = 400 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Silicon"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Silicon"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Silicon"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Silicon at T = 400K: " + str(Sol) + " cm^-3")


#T = 600
T = 600 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Silicon"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Silicon"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Silicon"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Silicon at T = 600K: " + str(Sol) + " cm^-3")


#Germanium

#T = 200
T = 200 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Germanium"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Germanium"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Germanium"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Germanium at T = 200K: " + str(Sol) + " cm^-3")


#T = 400
T = 400 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Germanium"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Germanium"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Germanium"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Germanium at T = 400K: " + str(Sol) + " cm^-3")


#T = 600
T = 600 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Germanium"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Germanium"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Germanium"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Germanium at T = 600K: " + str(Sol) + " cm^-3")




#Gallium Arsenide

#T = 200
T = 200 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Gallium Arsenide"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Gallium Arsenide"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Gallium Arsenide"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Gallium Arsenide at T = 200K: " + str(Sol) + " cm^-3")


#T = 400
T = 400 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Gallium Arsenide"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Gallium Arsenide"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Gallium Arsenide"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Gallium Arsenide at T = 400K: " + str(Sol) + " cm^-3")


#T = 600
T = 600 #K

Equation = IntrinsicCarrierConcentrationInASemiconductorEq

#Plug in Givens
Equation = Equation.subs(N_c, EffectiveDoS["Gallium Arsenide"]["N_c"]*(T/300)**1.5)
Equation = Equation.subs(N_v, EffectiveDoS["Gallium Arsenide"]["N_v"]*(T/300)**1.5)
Equation = Equation.subs(E_g, EffectiveDoS["Gallium Arsenide"]["E_g"])
Equation = Equation.subs("k", k)
Equation = Equation.subs("T", T/300)

#Solve for n_i
Sol = solve(Equation, n_i)[0]

#Print
print("Gallium Arsenide at T = 600K: " + str(Sol) + " cm^-3")


```
